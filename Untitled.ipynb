```python
from shioaji.data import Kbars
import pandas as pd
from datetime import datetime
from datetime import timedelta
import matplotlib.pyplot as plt
import time
```

```python
kbars = api.kbars(api.Contracts.Indexs['001'], start="2021-01-01", end="2021-10-15")
df = pd.DataFrame({**kbars})
df.ts = pd.to_datetime(df.ts)
df['tss'] = pd.to_datetime(df.ts)
```

```python
df.ts = df.ts.dt.date
df.ts = pd.to_datetime(df.ts)
```

```python
date1 = datetime.strptime("2007-01-01","%Y-%m-%d")
date2 = datetime.strptime("2007-12-31","%Y-%m-%d")
day=timedelta(days=1)
year=timedelta(days=365)
```

```python
groups = df.groupby(df.ts)
date3=date1
count=0;
plt.figure(figsize=(100,20))
while date1 <= date2:
    try:
        if count==3:
            plt.savefig(datetime.strftime(date3,"%Y-%m-%d")+"-"+datetime.strftime(date1,"%Y-%m-%d")+".png")
            date3=date1
            count=0
            plt.clf()
        df2=groups.get_group(date1)
        #if date1 != datetime.strptime("2019-12-20","%Y-%m-%d"):
        #    df2=df2.iloc[:int(df2.shape[0]/2),:]
        x = df2['tss'].index
        y = df2['Close']
        plt.plot(x, y)
        plt.title(datetime.strftime(date3,"%Y-%m-%d")+"-"+datetime.strftime(date1,"%Y-%m-%d"))
        plt.xlabel("time")
        plt.ylabel("price")
        date1 = date1 + day
        count = count + 1
    except:
        date1 = date1 + day
```

```python
while date1 <= date2:
    try:
        ticks = api.ticks(
            contract=api.Contracts.Indexs["001"], 
            date=datetime.strftime(date1,"%Y-%m-%d")
        )
        df = pd.DataFrame({**ticks})
        df.ts = pd.to_datetime(df.ts)
        df.to_csv(datetime.strftime(date1,"%Y-%m-%d")+".csv")
        date1 = date1 + day
    except:
        date1 = date1 + day
```
